# Datos de Diciembre

```python
import pandas as pd
```

```python
df_clientes = pd.read_csv('clientes_diciembre.csv',sep ='|')

df_productos = pd.read_csv('productos_diciembre.csv',sep='|')

df_finan = pd.read_csv('financiacion_diciembre.csv',sep='|')

df_consumo = pd.read_csv('consumos_diciembre.csv',sep='|')
```

```python
print(df_clientes['id'].isin(df_productos['id']).value_counts())
print(df_clientes['id'].isin(df_consumo['id']).value_counts())
print(df_clientes['id'].isin(df_finan['id']).value_counts())
```

```python
print(len(df_clientes['id'].unique()))

print(len(df_finan['id'].unique()))

print(len(df_productos['id'].unique()))

print(len(df_consumo['id'].unique()))
```

Si en el parámetro on no le meto nada, hace el join inner. En este caso como los id son únicos y todos los archivos tienen los mismos, da igual el parámetro how

```python
df_total_diciembre = pd.merge(df_clientes,df_productos, on=['id'], how='left')

df_total_diciembre = pd.merge(df_total_diciembre,df_finan, on=['id'], how='left')

df_total_diciembre = pd.merge(df_total_diciembre,df_consumo, on=['id'], how='left')
df_total_diciembre
```

# Datos de Enero

```python
df_clientes = pd.read_csv('clientes_enero.csv', sep='|')
df_productos = pd.read_csv('productos_enero.csv', sep='|')
df_finan = pd.read_csv('financiacion_enero.csv', sep='|')
df_consumo = pd.read_csv('consumos_enero.csv', sep='|')
```

```python
print(df_clientes['id'].isin(df_productos['id']).value_counts())
print(df_clientes['id'].isin(df_finan['id']).value_counts())
print(df_clientes['id'].isin(df_consumo['id']).value_counts())
```

```python
df_total_enero = pd.merge(df_clientes,df_productos, on=['id'], how='left')

df_total_enero = pd.merge(df_total_enero,df_finan, on=['id'], how='left')

df_total_enero = pd.merge(df_total_enero,df_consumo, on=['id'], how='left')

df_total_enero
```

```python
del df_clientes
del df_productos
del df_finan
del df_consumo
```
